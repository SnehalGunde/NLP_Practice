```python
#library to read docx
import docx2txt
```

```python
#reading job description
job_description=docx2txt.process('JobDescription.docx')


#reading resume
resume=docx2txt.process('Resume.docx')
```

```python
print(job_description)
```

```python
print(resume)
```

```python
#create a list to compare both of them
content=[job_description,resume]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#create object of countvetorizer
cv = CountVectorizer()

#apply fit transform to content
matrix = cv.fit_transform(content)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#creating similarity matrix
similarity_matrix= cosine_similarity(matrix)
```

```python
print(similarity_matrix)
```

```python
print("Resume Matches by : " + str(similarity_matrix[1][0]*100) + "%")
```
